```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
coimbra = pd.read_csv('dataR2_coimbra.csv')
coimbra
```

```python
c_patient = coimbra.loc[coimbra['Classification'] == 2]
c_control = coimbra.loc[coimbra['Classification'] == 1]
c_patient
```

```python
# group control table by age, take means
c_c_age = c_control.groupby(['Age']).mean()

c_c_age.reset_index(inplace=True)

c_c_age.drop(labels='Classification', axis=1, inplace=True)

c_c_age.index.names=['ID']

c_c_age = c_c_age.round(3)

# group patient table by age, take means
c_age = c_patient.groupby(['Age']).mean()

c_age.reset_index(inplace=True)

c_age.drop(labels='Classification', axis=1, inplace=True)

c_age.index.names=['ID']

c_age = c_age.round(3)

# export tables to csv files
c_c_age.to_csv('coimbra_control_age_means.csv')

c_age.to_csv('coimbra_patient_age_means.csv')
c_age
```
